```python
import matplotlib.pyplot as plt
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales")
plt.show()
```

```python
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python

df.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("Category")["Sales"].sum().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Category Wise Sales")
plt.ylabel("")
plt.show()
```

```python
print("Total Sales =", df["Sales"].sum())
print("Average Sales =", df["Sales"].mean())
print("Maximum Sale =", df["Sales"].max())
print("Minimum Sale =", df["Sales"].min())
```
